## Problem
The day when chestnut buns cover the solar system

I research that the diameter of the solar system is 287.46 billion km. Link: https://www.universetoday.com/15585/diameter-of-the-solar-system/

```python
import math
import numpy as np
```

```python
pi = math.pi
solar_diameter = 2.8746*(10**14)
```

```python
solar_volume = (1/6)*pi*(solar_diameter**3)
```

```python
chestnut_bun_volume = 10
```

```python
def compute_time(chestnut_bun_volume, solar_volume):
    count_time = 0
    time_list = [count_time]
    chestnut_bun_list = [chestnut_bun_volume]
    while chestnut_bun_volume < solar_volume:
        count_time += 5
        chestnut_bun_volume *= 2
        time_list.append(count_time)
        chestnut_bun_list.append(chestnut_bun_volume)
    return count_time, time_list, chestnut_bun_list
```

```python
result, time_list, chestnut_bun_list = compute_time(chestnut_bun_volume, solar_volume)
print("With chestnut bun volume is {}, the time it covers the solar system is {} minutes".format(chestnut_bun_volume, result))
```

## Display the graph

```python
"""
Display the graph.
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("The time chestnut bun cover the solar system")
plt.xlabel("Time[minute]")
plt.ylabel("Chestnut bun volume [m^3]")
plt.plot(time_list, chestnut_bun_list) 
plt.show()
```
